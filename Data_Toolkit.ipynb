**2. How does broadcasting work in NumPy?**

ANS:- Broadcasting in NumPy allows operations on arrays of different shapes by automatically expanding them so that they are compatible for element-wise operations, without copying data. This makes array computations more flexible and efficient.

 **Broadcasting Rules**

When operating on two arrays, NumPy compares their shapes element-wise from right to left:

1.If dimensions are equal, they're compatible.

2.If one of the dimensions is 1, it's "stretched" to match the other.

3.If dimensions are not equal and neither is 1, an error is raised.


 Compatible Shapes – Examples

Example 1: Scalar with Array

````python
import numpy as np

a = np.array([1, 2, 3])
b = 10
print(a + b)
````

````python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.array([10, 20, 30])
print(a + b)
````

 Incompatible Shapes – Example

````python
a = np.array([[1, 2, 3]])
b = np.array([[1], [2]])
print(a + b)  # Shape mismatch: (1, 3) and (2, 1)
````

**4. Explain the use of the groupby() method in Pandas?**

ANS:- The groupby() method in Pandas is used to group data based on one or more columns, and then apply aggregation, transformation, or filtering operations on those groups.

**Basic Idea**

Think of groupby() as:
"Split → Apply → Combine"

1.Split the data into groups

2.Apply a function (e.g., sum, mean, count) to each group

3.Combine the results into a new DataFrame or Series

````python
#Example
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [50000, 60000, 52000, 70000, 62000, 71000]
}

df = pd.DataFrame(data)

# Group by Department and calculate average salary
result = df.groupby('Department')['Salary'].mean()
print(result)
````

````python
#Multiple Aggregations
df.groupby('Department')['Salary'].agg(['mean', 'sum', 'max'])
````

**7. What is a heatmap, and when should it be used?**

ANS:- A **heatmap** is a **data visualization tool** that represents values in a **matrix or table format** using **colors**. It helps to quickly spot patterns, correlations, and outliers in data.


###  What Does a Heatmap Show?

* Each **cell** in a heatmap corresponds to a value.
* The **color** of the cell reflects the **magnitude** of the value.

  * Darker or more intense colors often represent higher values.
  * Lighter or cooler colors represent lower values.


###  Common Uses of Heatmaps

| Use Case               | Description                                                           |
| ---------------------- | --------------------------------------------------------------------- |
| **Correlation Matrix** | Show relationships between variables (e.g., correlation coefficients) |
| **Confusion Matrix**   | Visualize model performance in classification problems                |
| **Feature Importance** | Display relative importance of variables in models                    |
| **Pivot Tables**       | Analyze aggregated data (e.g., sales by region/month)                 |
| **Sensor/Grid Data**   | Represent spatial or geographic data (e.g., temperature over an area) |

---

###  Example: Correlation Heatmap in Python




````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset('iris')

# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
````

**8. What does the term “vectorized operation” mean in NumPy?**

ANS:- In **NumPy**, a **vectorized operation** refers to performing operations on **entire arrays (vectors, matrices, etc.) without writing explicit loops**. These operations are applied **element-wise** and are implemented in optimized C code behind the scenes—making them **much faster** than using Python loops.

---

###  Key Idea

> Instead of looping through elements one by one, NumPy applies the operation to all elements **at once**, using efficient low-level routines.


###  Example: Without and With Vectorization

####  Without Vectorization (Using Loop)





````python
a = [1, 2, 3]
b = [4, 5, 6]
c = [a[i] + b[i] for i in range(len(a))]
print(c)  # Output: [5, 7, 9]
````

**With Vectorization (Using NumPy)**

````python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a + b
print(c)  # Output: [5 7 9]
````

**13. Why is handling missing data important in Pandas?**

ANS:- Handling missing data is **crucial in Pandas** (and data analysis in general) because **missing values can distort insights, lead to incorrect conclusions, and break machine learning models**.

---

### **Why Handling Missing Data Is Important**

1.  Incomplete or Inaccurate Analysis**:

   * Missing values can skew **mean, median, correlations, and distributions**.
   * Can lead to false assumptions about the data.

2.  Errors in Computation**:

   * Many functions in Pandas, NumPy, or Scikit-learn don’t work properly if there are `NaN` values.
   * ML algorithms often throw errors when encountering missing data.

3.  Misleading Model Performance**:

   * Ignoring missing data can lead to **biased models** or models that **overfit** or **underperform**.

4.  Visualization Issues**:

   * Charts and plots may be incomplete or misleading if they include missing values.

---

###  **Common Techniques in Pandas**

Here’s how missing data is often handled:

| Method             | Description                                                                                          |
| ------------------ | ---------------------------------------------------------------------------------------------------- |
| `df.isnull()`      | Identifies missing values                                                                            |
| `df.dropna()`      | Drops rows/columns with missing values                                                               |
| `df.fillna(value)` | Fills missing values with a specified value (mean, median, etc.)                                     |
| Imputation         | Replacing missing values with statistical estimates (e.g., interpolation or using model predictions) |

---

###  **Example**



````python
import pandas as pd
df = pd.DataFrame({
    'Age': [25, None, 30, 22],
    'Score': [88, 90, None, 85]
})

# Fill missing age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Drop rows with any missing data
df.dropna(inplace=True)
````

**15. How does NumPy handle multidimensional arrays?**

ANS:-  NumPy handles multidimensional arrays using a powerful data structure called the ndarray (n-dimensional array). This allows you to efficiently work with data in 1D, 2D, 3D, or even higher dimensions using consistent and optimized operations.

** Key Ways NumPy Handles Multidimensional Arrays**

**1. Creating Multidimensional Arrays**

````python
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])

b = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
````

**2. Shape and Dimensions**

**.shape** tells you the dimensions (rows, columns, etc.)

**.ndim** gives the number of dimensions

**.size** gives the total number of elements

````python
a.shape
a.ndim
a.size
````

**3. Indexing and Slicing**

Just like Python lists, but extended to multiple dimensions:

````python
a[0, 1]     # Access element at first row, second column
b[1, 0, 1]  # Access nested element in a 3D array
````

**4. Broadcasting**

NumPy can perform operations on arrays of different shapes, automatically expanding them to match dimensions when possible.

````python
a = np.array([[1, 2], [3, 4]])
b = np.array([10, 20])

# b is broadcast to match a's shape
a + b  # Output: [[11, 22], [13, 24]]
````

**5. Vectorized Operations**

Operations on entire arrays are done element-wise and are highly efficient:

````python
a * 2       # Multiplies every element by 2
a + a       # Element-wise addition
np.sqrt(a)  # Square root of each element
````

**6. Reshaping and Transposing**

**reshape()** changes the shape without changing data

**transpose()** switches axes

````python
a.reshape(3, 2)
a.T  # Transpose
````

**7. Axis-Based Operations**

Functions like sum(), mean(), max() can operate along specified axes:

````python
np.sum(a, axis=0)  # Sum down columns
np.sum(a, axis=1)  # Sum across rows
````

ANS:- NumPy offers several **advanced features** that go far beyond basic array manipulation. These features make NumPy a **powerful tool for scientific computing, data analysis, and machine learning**. Here's a breakdown:

---

###  **Advanced Features of NumPy**

---

### 1. **Broadcasting**

* Allows arithmetic operations between arrays of **different shapes**.
* NumPy automatically expands dimensions to make shapes compatible.

```python
import numpy as np
a = np.array([[1], [2], [3]])  # Shape (3, 1)
b = np.array([10, 20, 30])     # Shape (3,)
result = a + b  # Shape (3, 3)
```

---

### 2. **Vectorization**

* Replaces slow Python loops with **fast array operations**.
* Speeds up numerical computations significantly.

```python
a = np.arange(1_000_000)

```

---

### 3. **Masked Arrays**

* Use `np.ma` to handle **invalid or missing values** in computations.

```python
import numpy.ma as ma
data = ma.array([1, 2, 3, -999], mask=[0, 0, 0, 1])
data.mean()  # Ignores the masked value
```

---

### 4. **Memory Mapping of Arrays**

* Allows working with **very large datasets** that don’t fit into memory.

```python
np.memmap('data.dat', dtype='float32', mode='w+', shape=(10000, 10000))
```

---

### 5. **Structured Arrays (Record Arrays)**

* Arrays that behave like tables with **named fields** (like a lightweight DataFrame).

```python
data = np.array([(1, 'Alice'), (2, 'Bob')], dtype=[('id', 'i4'), ('name', 'U10')])

---

### 6. **Advanced Indexing and Slicing**

---

### 8. **Fourier Transform and Signal Processing**

### 9. **Random Sampling and Distributions**

### 10. **Multidimensional Statistics**

---

###  **Key Features Pandas Offers for Time Series Analysis**

---

### 1. **DatetimeIndex and Time-Based Indexing**

* Pandas lets you use **datetime objects as indices**, allowing you to slice, filter, and align time-based data effortlessly.

```python

### 2. **Built-In Date Parsing**

pd.to_datetime('2025-05-19')  # Returns a Timestamp object

```python

Common frequencies include:

* `'D'` – day
* `'M'` – month
* `'Y'` – year

* Apply moving averages, rolling sums, etc.

```python

**20.What is the role of a pivot table in Pandas?**

ANS:- The **pivot table** in Pandas plays a crucial role in **summarizing, analyzing, and reorganizing** tabular data. It’s similar to Excel's pivot tables and is especially useful for:

* Aggregating data
* Creating multi-dimensional summaries
* Making data easier to analyze or visualize

---

###  **What Is a Pivot Table?**

A **pivot table** reshapes data by **grouping it on one or more keys**, and **applying an aggregation function** (like sum, mean, count, etc.) to the grouped data.

---

###  **Pandas Function: `pivot_table()`**

```python
pd.pivot_table(data,
               values='column_to_aggregate',
               index='row_grouping_column',
               columns='column_grouping_column',
               aggfunc='mean')  # or 'sum', 'count', etc.
```

---

###  **Key Roles of a Pivot Table**

| Role                           | Description                                                                       |
| ------------------------------ | --------------------------------------------------------------------------------- |
| **Aggregation**                | Calculates summaries (sum, average, count, etc.) for combinations of rows/columns |
| **Restructuring**              | Transforms long-format data into a more readable wide-format                      |
| **Multi-dimensional analysis** | Allows comparing multiple variables across multiple dimensions                    |
| **Data cleaning**              | Helps identify trends, outliers, or missing values easily                         |

---

**21.Why is NumPy’s array slicing faster than Python’s list slicing?**

ANS:- NumPy’s array slicing is faster than Python’s list slicing mainly because of how they are implemented and how data is stored and accessed under the hood.

---

###  **Reasons Why NumPy Array Slicing is Faster**

---

### 1. **Contiguous Memory Storage**

* **NumPy arrays** store data in a **contiguous block of memory**, meaning all elements are laid out sequentially.
* This allows **fast, low-level access** to slices without creating copies (slices are views, not copies).
* **Python lists**, however, are arrays of pointers to objects scattered in memory, so slicing requires creating a new list and copying references.

---

### 2. **Views vs Copies**

* **NumPy slicing returns a *view*** of the original array (no new data is copied, just a new way to look at the same data).
* **Python list slicing creates a *new list*** with references to the sliced elements.
* This means NumPy avoids the overhead of allocating new memory and copying data.

---

### 3. **Fixed Data Type & Homogeneity**

* NumPy arrays have a **fixed, homogeneous data type** (e.g., all floats or all ints), enabling **optimized, vectorized operations**.
* Python lists can contain heterogeneous objects, so slicing must handle references and type checks.

---

### 4. **Optimized C Implementations**

* NumPy’s slicing operations are implemented in **optimized C code**.
* Python list slicing is implemented at the Python interpreter level, which is slower.
